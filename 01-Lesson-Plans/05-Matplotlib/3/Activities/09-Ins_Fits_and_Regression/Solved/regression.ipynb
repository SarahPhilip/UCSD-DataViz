```python
# Dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
# Set data
x_axis = range(10)
fake = [1, 2.5, 2.75, 4.25, 5.5, 6, 7.25, 8, 8.75, 9.8]
```

```python
# Set line
slope, intercept, _, _, _ = linregress(x_axis, fake)
fit = slope * x_axis + intercept
```

```python
# Plot data
plt.plot(x_axis, fake, linewidth=0, marker='o')
plt.plot(x_axis, fit, 'b--')

plt.xlim(0, 10)
plt.ylim(0, 10)

plt.title("Fake Banana Data!", fontsize=16, fontweight="bold")
plt.xlabel("Fake Banana Ages (in days)")
plt.ylabel("Fake Banana Weights (in Hundres of Kilograms)")
```
