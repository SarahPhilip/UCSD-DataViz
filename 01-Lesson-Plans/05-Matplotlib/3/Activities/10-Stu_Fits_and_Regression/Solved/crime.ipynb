```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load data
crime_data = pd.read_csv("../Resources/crime_data.csv")

year = crime_data['year']
violent_crime_rate = crime_data['violent crime rate']
murder_rate = crime_data[' murder and non-negligent manslaughter rate']  # column has space at start
aggravated_assault_rate = crime_data['aggravated assault rate']
```

```python
# Calculate violent crime regression
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(year, violent_crime_rate)
vc_fit = vc_slope * year + vc_int
```

```python
# Calculate murder rate regression
m_slope, m_int, m_r, m_p, m_std_err = stats.linregress(year, murder_rate)
m_fit = m_slope * year + m_int
```

```python
# Calculate aggravated assault regression
aa_slope, aa_int, aa_r, aa_p, aa_std_err = stats.linregress(year, aggravated_assault_rate)
aa_fit = aa_slope * year + aa_int
```

```python
# Plot violent crime
plt.figure(figsize=(6, 2))
plt.plot(year, violent_crime_rate, linewidth=1, marker="o", color="b")
plt.plot(year, vc_fit, "b--", linewidth=1)
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()

# Plot murder
plt.figure(figsize=(6, 2))
plt.plot(year, murder_rate, linewidth=1, marker="o", color="r")
plt.plot(year, m_fit, "r--", linewidth=1)
plt.xlabel('Year')
plt.ylabel('Murder Rate')
plt.show()

# Plot aggravated assault
plt.figure(figsize=(6, 2))
plt.plot(year, aggravated_assault_rate, linewidth=1, marker="o", color="g")
plt.plot(year, aa_fit, "g--", linewidth=1)
plt.xlabel('Year')
plt.ylabel('Aggrevated Assault Rate')
plt.show()
```

```python
# Print results and save image
year = 2019
print("The violent crime rate in 2019 will be " + str(vc_slope * year + vc_int) + ".")
print("The murder rate in 2019 will be " + str(m_slope * year + m_int) + ".")
print("The aggravated assault rate in 2019 will be " + str(aa_slope * year + aa_int) + ".")
```
