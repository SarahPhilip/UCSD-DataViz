```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Read in data
general_data = pd.read_csv("../Resources/general_heights.csv")
wba_data = pd.read_csv("../Resources/wba_data.csv")

general_heights = general_data['height']
wba_heights = wba_data['height']
```

```python
# Run the t-test
t_stat, p = stats.ttest_ind(general_heights, wba_heights, equal_var=False)
```

```python
# Report the data
print("The mean height of WBA players is {}.".format(wba_heights.mean()))
print("The mean height of women sampled is {}.".format(general_heights.values.mean()))
print("p is {}.".format(p))

if p < 0.05:
    print("The difference in sample means is significant.")
else:
    print("The difference in sample means is not significant.")
```

```python
# Plot sample means with error bars
tick_labels = ["General Public", "WBA Players"]

means = [general_heights.mean(), wba_heights.mean()]
sem = [general_heights.sem(), wba_heights.sem()]
```

```python
# Plot mean height of players
plt.errorbar(tick_labels, means, yerr=sem, fmt="o")

plt.xlim(-0.5, 1.5)
plt.title("Mean Height of Women in General Population and WBA Players")
plt.ylabel("Height (Inches)")
```
