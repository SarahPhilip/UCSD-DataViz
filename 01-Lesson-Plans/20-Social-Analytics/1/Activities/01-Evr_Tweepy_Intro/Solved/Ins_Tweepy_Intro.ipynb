```python
import json
import tweepy
from pprint import pprint
```

```python
# Import Twitter API Keys
from config import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Get all tweets from followed users
for tweet in api.home_timeline():
    print('-' * 80)
    pprint(tweet)
```

```python
# Get own tweets
for tweet in api.user_timeline():
    print('-' * 80)
    pprint(tweet)
```

```python
# Get tweets for a user
target_user = "HamillHimself"

# Loop through all tweets
for tweet in api.user_timeline(target_user):
    print('-' * 80)
    pprint(tweet)
```

```python
# Get single tweet data
target_tweets = api.user_timeline(target_user)
first_tweet = target_tweets[0]
pprint(first_tweet)
```

```python
# Get the tweet's user, number of followers, and number of tweets
print('User:', first_tweet['user']['name'])
print('Followers:', first_tweet['user']['followers_count'])
print('Tweets:', first_tweet['user']['statuses_count'])
```

```python
# Get the tweet's text
print('Latest Tweet:', first_tweet['text'])
```
