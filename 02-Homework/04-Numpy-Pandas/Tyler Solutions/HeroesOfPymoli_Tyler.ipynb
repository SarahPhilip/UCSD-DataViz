## Heroes Of Pymoli Data Analysis

```python
import pandas as pd
import numpy as np

# Keystroke saver :)
as_currency = '${:,.2f}'.format

# Load CSV
purchase_df = pd.read_csv('Resources/purchase_data.csv')
purchase_df.head()
```

## Player Count

```python
# Track total unique players (used later)
total_unique_players = purchase_df['SN'].nunique()

# Output player count
pd.DataFrame({
    'Total Players': [total_unique_players],
})
```

## Purchasing Analysis (Total)

```python
# Output purchase analysis
pd.DataFrame({
    'Number of Unique Items': [purchase_df['Item ID'].nunique()],
    'Total Revenue': [as_currency(purchase_df['Price'].sum())],
    'Number of Purchases': [purchase_df['Price'].count()],
    'Average Price': [as_currency(purchase_df['Price'].mean())],
})
```

## Gender Demographics

```python
# Group data by player
player_group = purchase_df.groupby('SN')

player_df = pd.DataFrame({
    'Age': player_group['Age'].first(),
    'Gender': player_group['Gender'].first(),
    'Purchase Count': player_group['Price'].count(),
    'Total Purchase Value': player_group['Price'].sum(),
})

# Uncomment to test/view output
# player_df.head()
```

```python
# Group player data by gender
player_gender_group = player_df.groupby('Gender')

# Count players by gender
players_by_gender = player_gender_group['Gender'].count()

# Output gender stats
pd.DataFrame({
    'Percentage of Players': (100 * players_by_gender / total_unique_players).round(2),
    'Total Count': players_by_gender,
})
```


## Purchasing Analysis (Gender)

```python
# Group purchase data by gender
purchase_gender_group = purchase_df.groupby('Gender')

# Output gender purchase stats
pd.DataFrame({
    'Purchase Count': purchase_gender_group['Price'].count(),
    'Average Purchase Price': purchase_gender_group['Price'].mean().map(as_currency),
    'Total Purchase Value': purchase_gender_group['Price'].sum().map(as_currency),
    'Avg Total Purchase per Person': player_gender_group['Total Purchase Value'].mean().map(as_currency),
})
```

## Age Demographics

```python
# Add Age Range bin to player_df
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 99999]
age_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
player_df['Age Range'] = pd.cut(player_df['Age'], age_bins, labels=age_labels, right=False)

# Group player data by Age Range
player_age_range_group = player_df.groupby('Age Range')

# Count players by Age Range
players_by_age_range = player_age_range_group['Age Range'].count()

# Output age demographics
pd.DataFrame({
    'Percentage of Players': (100 * players_by_age_range / total_unique_players).round(2),
    'Total Count': players_by_age_range,
})
```

## Purchasing Analysis (Age)

```python
# Add Age Range bin to purchase_df
purchase_df['Age Range'] = pd.cut(purchase_df['Age'], age_bins, labels=age_labels, right=False)

# Group purchase data by Age Range
purchase_age_range_group = purchase_df.groupby('Age Range')

# Output gender purchase stats
pd.DataFrame({
    'Purchase Count': purchase_age_range_group['Price'].count(),
    'Average Purchase Price': purchase_age_range_group['Price'].mean().map(as_currency),
    'Total Purchase Value': purchase_age_range_group['Price'].sum().map(as_currency),
    'Avg Total Purchase per Person': player_age_range_group['Total Purchase Value'].mean().map(as_currency),
})
```

## Top Spenders

```python
# Create top_spenders_df, ordered by descending total purchase
top_spenders_df = pd.DataFrame({
    'Purchase Count': player_df['Purchase Count'],
    'Average Purchase Price': player_df['Total Purchase Value'] / player_df['Purchase Count'],
    'Total Purchase Value': player_df['Total Purchase Value'],
}).sort_values('Total Purchase Value', ascending=False)

# Add currency formatting
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map(as_currency)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map(as_currency)

# Output top 5 spenders
top_spenders_df.head()
```

## Most Popular Items

```python
# Group purchases by items
purchase_item_group = purchase_df.groupby(['Item ID', 'Item Name'])

# Create popular_items_df
purchase_item_df = pd.DataFrame({
    'Purchase Count': purchase_item_group['Price'].count(),
    'Item Price': purchase_item_group['Price'].first(),
    'Total Purchase Value': purchase_item_group['Price'].sum(),
})

# Sort by number of purchase
popular_items_df = purchase_item_df.sort_values('Purchase Count', ascending=False)

# Add currency formatting
popular_items_df['Item Price'] = popular_items_df['Item Price'].map(as_currency)
popular_items_df['Total Purchase Value'] = popular_items_df['Total Purchase Value'].map(as_currency)

# Output Top 5 items
popular_items_df.head()
```

## Most Profitable Items

```python
# Sort by number of purchase
profitable_items_df = purchase_item_df.sort_values('Total Purchase Value', ascending=False)

# Add currency formatting
profitable_items_df['Item Price'] = profitable_items_df['Item Price'].map(as_currency)
profitable_items_df['Total Purchase Value'] = profitable_items_df['Total Purchase Value'].map(as_currency)

# Output Top 5 items
profitable_items_df.head()
```
